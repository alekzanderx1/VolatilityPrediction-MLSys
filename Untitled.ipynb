```python
ff = pd.read_csv("FammaFrench.csv")
ff.rename(columns = {',Mkt-RF,SMB,HML,RF': 'data'}, inplace = True)
ff.drop(index = len(ff) -1, inplace = True)
```

```python
data = []
for index, row in ff.iterrows():
    raw = row['data'].split(',')
    time = dt.strptime(raw[0], '%Y%m%d')
    excess = float(raw[1])
    smb = float(raw[2])
    hml = float(raw[3])
    rf = float(raw[4])
    data.append([time, excess, smb, hml, rf])
    
```
